Discover the input for a graph ***G(V,E)*** so that the count of the **cut- edges will be 3** where vertices ***V*** and edges ***E*** are:

 - V: (0, 1, 2, 3) 
 - E: (0, 1), (0, 2), (0, 3)

by using Grover's Algorithm with **iteration = 2** times.

**Expected Outputs:**
 - Quantum circuit.
 - The quantum states (bit strings separated by commas) with the highest probabilities.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit import IBMQ, Aer, execute
from qiskit.tools.visualization import plot_histogram
import math
```

```python
V = (0, 1, 2, 3)
E = (0, 1), (0, 2), (0, 3)

v = len(V)
e = len(E)

q = QuantumRegister(2*e+3)
c = ClassicalRegister(v)
qc = QuantumCircuit(q, c)

# cutedge checker
def ccheck(a, b, c):
    qc.cx(q[a], q[c])
    qc.cx(q[b], q[c])

# inverse operations
def iccheck(a, b, c):
    qc.cx(q[b], q[c])
    qc.cx(q[a], q[c])


# initialization
qc.h(q[0:v])
qc.x(q[2*e+2])
qc.h(q[2*e+2])
qc.barrier()

for i in range(math.floor((math.pi/4) * (math.sqrt(2 ** e)))):
    # oracle
    for count, edge in enumerate(E):
        ccheck(edge[0], edge[1], e+count+1)

    qc.mct([q[e+i+1] for i in range(e)], q[-1])
    # inverse
    for count, edge in reversed(list(enumerate(E))):
        iccheck(edge[0], edge[1], e+count+1)

    qc.barrier()

    # diffusion
    qc.h(q[0:v])
    qc.barrier()
    qc.x(q[0:v])
    qc.barrier()
    qc.h(q[e])
    qc.mct([q[i] for i in range(e)], q[-1])
    qc.h(q[e])
    qc.barrier()
    qc.x(q[0:v])
    qc.barrier()
    qc.h(q[0:v])
    qc.barrier()

qc.measure(q[0:v], c[0:v])
qc.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=10000)
result = job.result()
count = result.get_counts()
print(count)
plot_histogram(count)
```
